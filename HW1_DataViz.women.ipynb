```python
import pandas as pd
import matplotlib.pyplot as plt

files = {
   "1994-1998": "1994-1998.csv",
   "2003-2004": "2003-2004.csv",
   "2005-2006": "2005-2006.csv",
   "2007-2008": "2007-2008.csv"
}

totals = ["Total fruit", "Vegetables, Total", "Dairy, Total"]

dfs = []
for year, file in files.items():
    df = pd.read_csv(file)
    df = df[df['Food'].isin(totals)]
    df = df[['Food', 'Women']].copy()
    df['Year'] = year
    dfs.append(df)

data = pd.concat(dfs)

pivot = data.pivot(index="Year", columns="Food", values="Women")

pivot.plot(kind="bar", figsize=(8,6))
plt.title("Food Totals (Women)")
plt.xlabel("Year Range")
plt.ylabel("Consumption (%)")
plt.legend(title="Food")
plt.tight_layout()

plt.figtext(
    0.5, -0.08,
    "Source: Appendix B, Food At Home — USDA ERS (1994–2008) https://www.ers.usda.gov/data-products/food-availability-per-capita-data-system/",
    ha="center", fontsize=8
)

plt.savefig("prob2_women.png", dpi=300, bbox_inches="tight")
plt.show()


```
