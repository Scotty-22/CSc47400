```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

files_map = {
    "1994-1998": "1994-1998.csv",
    "2003-2004": "2003-2004.csv",
    "2005-2006": "2005-2006.csv",
    "2007-2008": "2007-2008.csv"
}

fruits = ["Apples as fruit", "Bananas", "Berries", "Grapes",
          "Melons", "Oranges, Total", "Other citrus fruit",
          "Stone fruit", "Tropical fruit"]

dfs = []
for year, file in files_map.items():
    df = pd.read_csv(file)
    df = df[df['Food'].isin(fruits)]
    df = df[['Food', 'Women']].copy()
    df['Year'] = year
    dfs.append(df)
data = pd.concat(dfs)

plt.figure(figsize=(12,6))
for item in fruits:
    subset = data[data['Food'] == item]
    plt.plot(subset['Year'], subset['Women'], marker='o', label=item)

plt.xlabel("Year Range")
plt.ylabel("Consumption (%)")
plt.title("Fruit Consumption (Women)")
plt.legend()
plt.tight_layout()
plt.savefig("fruits_women.png", dpi=300, bbox_inches="tight")
plt.show()

files.download("fruits_women.png")






```
