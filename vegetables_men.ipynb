```python
import pandas as pd
import matplotlib.pyplot as plt

files = {
    "1994-1998": "1994-1998.csv",
    "2003-2004": "2003-2004.csv",
    "2005-2006": "2005-2006.csv",
    "2007-2008": "2007-2008.csv"
}

vegetables = [
    "Vegetables, Total", "Brassica, Total", "Broccoli and cauliflower",
    "Carrots", "Celery", "Cucumbers", "Green Peas", "Leafy vegetable, Total",
    "Lettuce", "Onions", "Peppers", "Tomatoes", "Sweet corn",
    "Potatoes", "Snap beans", "Legumes dried"
]

dfs = []
for year, file in files.items():
    df = pd.read_csv(file)
    df = df[df['Food'].isin(vegetables)]
    df = df[['Food', 'Men']].copy()
    df['Year'] = year
    dfs.append(df)

data = pd.concat(dfs)

plt.figure(figsize=(12,6))
for item in vegetables:
    subset = data[data['Food'] == item]
    plt.plot(subset['Year'], subset['Men'], marker='o', label=item)

plt.xlabel("Year Range")
plt.ylabel("Consumption (%)")
plt.title("Vegetable Consumption (Men)")
plt.legend()
plt.tight_layout()
plt.figtext(
    0.5, -0.02,
    "Source: Appendix B, Food At Home  USDA ERS (19942008) https://www.ers.usda.gov/data-products/food-availability-per-capita-data-system/",
    ha="center", fontsize=8
)
plt.savefig("vegetables_men.png", dpi=300, bbox_inches="tight")
plt.show()



```
